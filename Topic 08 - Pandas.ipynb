# Introduction

Pandas is a numpy-based package widely used for data science. Here we will use a small fraction of its capabilities. The pandas syntax is a bit different from the one already used.

Pandas have a very extensive range of functions beyond what I will show here. The focus of this class is to open data provided by simulations in the format of `txt`,` csv`, `out`, etc.

# Creating a DataFrame


```python
import pandas as pd
import numpy as np
dataf = pd.DataFrame({'No': [50, 20,21], 'Yes': [131, 5, 2], 'Nothing': [0,0,0]},
            index=['product A', 'product B', 'product C'])

print(dataf)
print()
print(dataf.index)
print()
print(dataf.columns)
print()
print('Extracting a column:')
print(dataf.loc[:,'No'])
```

# Opening files

Pandas allows to open .csv .out .dat and .xls files easily. To do this, we use the **read_csv** or **read_excel** functions. The fisrt 5 rows of a DataFrame can be seen with the method **head** and the last 5 rows with the method **tail**.

```python
movieData = pd.read_csv('./Dataset/movie_metadata.csv')
movieData.head()
```

```python
movieData.tail()
```

Some useful data of the DaraFrame can be obtained with the method **info**

```python
movieData.info()
```

Note that in the previous datast there are some *NaN* or *missing values*. You can delete those values with the method *dropna()*

```python
movieDataNoMissing = movieData.dropna()
movieDataNoMissing.info()
```

Note that now all the fields have the same quantity of elements. If you want to perform a simple statistical analysis of the DataFrame you can use the method *describe()*.

```python
movieDataNoMissing.describe()
```

pandas has different options to open files with complex formats. You can explore all the options with *help(pd.read_csv)*

```python
help(pd.read_csv)
```

# Adding Columns
We can add new columns to DataFrames easily:

```python
dataf['New'] = dataf['No'] + 10

a  = dataf["New"]
print(dataf.head())
```

# Selecting datas from  Dataframes

To select data inside a DataFrame you can use one of the methods:

```python
df1 = dataf.iloc[:1, :]
df1.head()
```

```python
cols = ['No', 'Yes']
df1 = dataf[cols]
df1.head()
```

```python
dataf.loc[:, 'No':'Yes']
```

Note that the **loc** method also accepts column names.
